```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the data
df = pd.read_csv('/content/AviationData.csv', encoding='latin-1', low_memory=False)

# Convert Event.Date to datetime
df['Event.Date'] = pd.to_datetime(df['Event.Date'])
df['Year'] = df['Event.Date'].dt.year
df['Month'] = df['Event.Date'].dt.month
df['Hour'] = df['Event.Date'].dt.hour

# 1. Analyze incidents by year
yearly_incidents = df['Year'].value_counts().sort_index()
plt.figure(figsize=(15, 6))
yearly_incidents.plot(kind='line')
plt.title('Aviation Incidents by Year')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.grid(True)
plt.tight_layout()
plt.show()

# 3. Analyze incidents by aircraft damage
aircraft_damage = df['Aircraft.damage'].value_counts()
plt.figure(figsize=(10, 6))
aircraft_damage.plot(kind='bar')
plt.title('Incidents by Aircraft Damage')
plt.xlabel('Aircraft Damage')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 4. Analyze incidents by weather condition
weather_condition = df['Weather.Condition'].value_counts()
plt.figure(figsize=(10, 6))
weather_condition.plot(kind='bar')
plt.title('Incidents by Weather Condition')
plt.xlabel('Weather Condition')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 5. Analyze incidents by phase of flight
flight_phase = df['Broad.phase.of.flight'].value_counts()
plt.figure(figsize=(10, 6))
flight_phase.plot(kind='bar')
plt.title('Incidents by Phase of Flight')
plt.xlabel('Phase of Flight')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 6. Correlation analysis of numerical variables
numerical_columns = ['Number.of.Engines', 'Total.Fatal.Injuries', 'Total.Serious.Injuries', 'Total.Minor.Injuries', 'Total.Uninjured']
correlation_matrix = df[numerical_columns].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
plt.title('Correlation Matrix of Numerical Factors')
plt.tight_layout()
plt.show()

# 7. Top 10 countries with most incidents
top_countries = df['Country'].value_counts().head(10)
plt.figure(figsize=(12, 6))
top_countries.plot(kind='bar')
plt.title('Top 10 Countries with Most Aviation Incidents')
plt.xlabel('Country')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 8. Incidents by purpose of flight
purpose_of_flight = df['Purpose.of.flight'].value_counts()
plt.figure(figsize=(12, 6))
purpose_of_flight.plot(kind='bar')
plt.title('Incidents by Purpose of Flight')
plt.xlabel('Purpose of Flight')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 9. Incidents by aircraft category
aircraft_category = df['Aircraft.Category'].value_counts()
plt.figure(figsize=(12, 6))
aircraft_category.plot(kind='bar')
plt.title('Incidents by Aircraft Category')
plt.xlabel('Aircraft Category')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("Analysis complete. Please check the visualizations for insights.")
```
